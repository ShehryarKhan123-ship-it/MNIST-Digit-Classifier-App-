**Fetching Dataset of MNIST from sklearn built in dataste**

```python
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784',version=1)
mnist.keys()
```

 **Perfroming Train , test , Split on 60 to 10 ratio**


```python


from sklearn.model_selection import train_test_split

X, y = mnist["data"], mnist["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, train_size=0.6, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

```

**Random Forest Model Development and Classification**

```python

from sklearn.ensemble import RandomForestClassifier

rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rnd_clf.fit(X_train, y_train)


from sklearn.metrics import accuracy_score
y_pred = rnd_clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

**Classification_Report for Random Forest Model**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.imshow(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import f1_score ,precision_score, recall_score
print(f1_score(y_test,y_pred,average='macro'))
print(precision_score(y_test,y_pred,average='macro'))
print(recall_score(y_test,y_pred,average='macro'))
```

**Confusion Matrix  visualization Metrics**

```python
import numpy as np

confusion_matrix = confusion_matrix(y_test, y_pred)
row_sums = confusion_matrix.sum(axis=1, keepdims=True)
norm_conf_mx = confusion_matrix / row_sums
np.fill_diagonal(norm_conf_mx, 0)
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()
```

```python
import numpy as np

# Get the indices of the misclassified samples
misclassified_indices = np.where(y_pred != y_test)[0]

# Sort the misclassified samples by their prediction probabilities (confidence)
# in ascending order (least confident first)
# Assuming your model has a `predict_proba` method
# If not you need to modify the sorting criterion.
try:
    # Use .iloc to select rows using integer indices
    probabilities = rnd_clf.predict_proba(X_test.iloc[misclassified_indices])
    confidence = np.max(probabilities, axis=1) # confidence of the predicted class.
    sorted_indices = np.argsort(confidence) # indices to sort confidence scores
    misclassified_indices = misclassified_indices[sorted_indices] # sorted indices

    # Plot the worst misclassifications (least confident)
    num_worst = 9 # number of worst misclassifications to plot
    worst_indices = misclassified_indices[:num_worst]

    fig, axes = plt.subplots(3, 3, figsize=(8, 8))
    for i, ax in enumerate(axes.ravel()):
        # Use .iloc to access rows using integer location
        image = X_test.iloc[worst_indices[i]].values.reshape(28, 28)
        ax.imshow(image, cmap="gray")
        ax.set_title(f"Pred: {y_pred[worst_indices[i]]}, True: {y_test.iloc[worst_indices[i]]}")
        ax.axis("off")
    plt.tight_layout()
    plt.show()
except AttributeError:
    print("The classifier does not have a predict_proba method.")
    print("Plotting misclassified images in random order.")
    # Plot the first few misclassified samples if predict_proba is unavailable
    fig, axes = plt.subplots(3, 3, figsize=(8, 8))
    for i, ax in enumerate(axes.ravel()):
        if i < len(misclassified_indices):
          # Use .iloc for row selection
          image = X_test.iloc[misclassified_indices[i]].values.reshape(28, 28)
          ax.imshow(image, cmap="gray")
          ax.set_title(f"Pred: {y_pred[misclassified_indices[i]]}, True: {y_test.iloc[misclassified_indices[i]]}")
          ax.axis("off")
        else:
          ax.axis("off") # Hide empty subplots if less than 9 errors
    plt.tight_layout()
    plt.show()
```

**SGD Classifier model training and evaluation**

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score,precision_score

sgd_clf = SGDClassifier(loss="hinge", random_state=42)
sgd_clf.fit(X_train, y_train)

y_pred = sgd_clf.predict(X_test)
accuracy_score(y_test,y_pred)

print(classification_report(y_test,y_pred))

print(f1_score(y_test,y_pred,average='macro'))
print(precision_score(y_test,y_pred,average='macro'))
print(recall_score(y_test,y_pred,average='macro'))
```

**Optimizing SGDClassifier with GridSearchCV
Briefing:**

We employ GridSearchCV to systematically explore different combinations of hyperparameters for our SGDClassifier model. This process is crucial for finding the optimal settings that maximize the model's performance on the given dataset.

**Why GridSearchCV?**

Automated Hyperparameter Tuning: GridSearchCV automates the tedious task of manually trying different hyperparameter values. It exhaustively searches through a predefined grid of hyperparameters, evaluating each combination using cross-validation.
Performance Enhancement: By finding the best hyperparameter settings, GridSearchCV helps us significantly improve the model's accuracy and overall performance.
Efficient and Systematic: GridSearchCV provides a structured and efficient approach to hyperparameter optimization, ensuring that we thoroughly explore the hyperparameter space.
In our case, we are using GridSearchCV to find the best combination of loss, penalty, alpha, and max_iter for the SGDClassifier, ultimately aiming to achieve the highest possible accuracy on the MNIST dataset.

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Predict
y_pred = sgd_clf.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='OrRd')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - SGDClassifier')
plt.show()
```

**Misclassification Count Per Class**

```python
from collections import Counter

# Create mask for misclassified samples
misclassified = y_test != y_pred

# Count true labels of misclassified points
misclass_count = Counter(y_test[misclassified])

# Plot misclassified count per class
plt.figure(figsize=(6,4))
plt.bar(misclass_count.keys(), misclass_count.values(), color='crimson')
plt.title("Misclassifications per Class")
plt.xlabel("Class Label")
plt.ylabel("Count")
plt.show()
```

```python
import gradio as gr
import joblib
import numpy as np
from PIL import Image

# Load saved models
rnd_clf = joblib.load('rnd_clf.pkl')
sgd_clf = joblib.load('sgd_clf.pkl')

# Prediction function
def predict_digit(image):
    # Handle both PIL image and numpy array
    if isinstance(image, np.ndarray):
        image = Image.fromarray(image)

    image = image.convert('L').resize((28, 28))
    image = np.array(image)
    image = 255 - image  # Invert colors
    image = image / 255.0
    image = image.reshape(1, -1)

    sgd_pred = rnd_clf.predict(image)[0]
    rf_pred = sgd_clf.predict(image)[0]

    return f"SGDClassifier: {sgd_pred}, RandomForest: {rf_pred}"

# Launch Gradio interface
gr.Interface(
    fn=predict_digit,
    inputs=gr.Image(type="pil", label="Upload or Draw Digit"),
    outputs=gr.Textbox(label="Predicted Digit"),
    title="MNIST Digit Classifier",
    description="Upload or draw a digit (0-9) to get predictions from SGD and Random Forest."
).launch(share=True, pwa=True,debug=True)

```
